```python
import statistics  
  
class DataFrame():  
    def __init__(self, column, fill_value=0):  
        # Инициализируем атрибуты  
        self.column = column  
        self.fill_value = fill_value  
        # Заполним пропуски  
        self.fill_missed()  
        # Конвертируем все элементы в числа  
        self.to_float()  
          
    def fill_missed(self):  
        for i, value in enumerate(self.column):  
            if value is None or value == '':  
                self.column[i] = self.fill_value  
                  
    def to_float(self):  
        self.column = [float(value) for value in self.column]  
      
    def median(self):  
        return statistics.median(self.column)  
      
    def mean(self):  
        return statistics.mean(self.column)  
      
    def deviation(self):  
        return statistics.stdev(self.column)  
      
  
      
# Воспользуемся классом  
df = DataFrame(["1", 17, 4, None, 8])  
  
print(df.column)  
# => [1.0, 17.0, 4.0, 0.0, 8.0]  
print(df.deviation())  
# => 6.89  
print(df.median())  




class IntDataFrame():
    
    def __init__(self , column):
        self.column =list(map(int,column))  
        
    def count(self):
        counts = 0
        for i in self.column:
            if i >0:
                counts +=1
        return counts
        
        
    def unique(self):
        return len(set(self.column))
    


df = IntDataFrame([4.7, 4, 3, 0, 2.4, 0.3, 4])

print(df.column)
# [4, 4, 3, 0, 2, 0, 4]     
print(df.count())
# 5

print(df.unique())



class User():
    
    def __init__(self , email, password , balance):
        self.email = email
        self.password = password
        self.balance = balance
        
        
    def login(self , email , password):
        if self.email == email and self.password == password:
            return True
        else:
            return False
        
    def update_balance(self, amount):
        self.balance = self.balance + amount
        
```

```python
class OwnLogger():
    def __init__(self):
        self.logs = {"info": None, "warning": None, "error": None, "all": None}
        self.all_logs = []
        
    def log (self ,message , level):
        if level in self.logs:
            self.logs[level] = message
            self.all_logs.append((message,level))
            self.logs['all'] = message
        
    def show_last(self , level = 'all'):
        if level == ['all']:
            return self.logs['all']
        elif level in self.logs:
            return self.logs[level]
        else:
            return None
    
    
    
logger = OwnLogger()
logger.log("This is an info message", "info")
logger.log("This is a warning message", "warning")
logger.log("This is an error message", "error")

print(logger.show_last("info"))     # Вывод: This is an info message
print(logger.show_last("warning"))  # Вывод: This is a warning message
print(logger.show_last("error"))    # Вывод: This is an error message
print(logger.show_last("all"))      # Вывод: This is an error message
print(logger.show_last())           # Вывод: This is an error message
print(logger.show_last("debug"))    # Вывод: None
```

```python
class ParentException(Exception):
    def __init__(self,message , error):
        super().__init__(message)
        print(f'Errors : {error}')
        
        
class MyException(ParentException):
    def __init__(self, message, error):
        super().__init__(message, error)
        


MyException.__dict__
MyException.__doc__ = 'description for class'



try:
    raise MyException('message' , 'error')
except ParentException as e:
    print(e)
```
